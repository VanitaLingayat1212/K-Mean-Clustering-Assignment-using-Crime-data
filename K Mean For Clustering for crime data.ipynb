```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
crime_data = pd.read_csv(r'C:\Users\Vanita\OneDrive\Documents\ASSIGNMENTS EXCELR NEW\Clustering assignments\crime_data.csv')
```

```python
crime_data
```

```python
crime_data1 = crime_data.drop(['Unnamed: 0'], axis = 1)
```

```python
crime_data1
```

```python
crime_data1.info()
```

```python
crime_data1_norm=StandardScaler().fit_transform(crime_data1)
```

```python
crime_data1_norm
```

```python
wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,random_state=0)
    kmeans.fit(crime_data1_norm)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
clusters4=KMeans(4,random_state=40).fit(crime_data1_norm)
clusters4
```

```python
clusters4.labels_
```

```python
crime_data4=crime_data1.copy()
crime_data4['cluster4sid']=clusters4.labels_
crime_data4
```

```python
clusters4.cluster_centers_
```

```python
crime_data4.groupby('cluster4sid').agg(['mean']).reset_index()
```

```python
plt.figure(figsize=(12, 9))  
plt.scatter(crime_data4['cluster4sid'],crime_data4['UrbanPop'], c=clusters4.labels_)
```
